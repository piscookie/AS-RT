```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from matplotlib.colors import Normalize
import matplotlib.cm as cm

plt.rcParams['font.serif'] = ['Times New Roman'] 
plt.rcParams['font.size'] = 14

#  CSV 
df = pd.read_csv('imagemol_pretrain_feature/features_seed1.csv')

#  t-SNE  '0' 
X = df.iloc[:, df.columns.get_loc('0'):].values

#  'label' 
labels = df['label']  #  'label' 

#  t-SNE 
tsne = TSNE(n_components=2, perplexity=5, random_state=0)

#  t-SNE 
X_embedded = tsne.fit_transform(X)

# 
fig, ax = plt.subplots(figsize=(6, 4), dpi=150)
ax.set_facecolor('#F6FAFF')
cmap = cm.RdYlGn_r #
norm = Normalize(vmin=0, vmax=600)    

# 
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels,cmap=cmap, s=180, edgecolor='white', norm=norm)

# 
cbar = plt.colorbar(scatter)
cbar.set_label('')

plt.rcParams['font.serif'] = ['Times New Roman'] 
plt.rcParams['font.size'] = 12

# 
plt.title('Pretraining Features')
plt.xlabel('', fontsize=10)
plt.ylabel('', fontsize=10)

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from matplotlib.colors import Normalize
#  CSV 
df = pd.read_csv('imagemol_finturn_Features/SMILES-SEED1-reg.csv')

#  t-SNE  '0' 
X = df.iloc[:, df.columns.get_loc('0'):].values

#  'label' 
labels = df['label']  #  'label' 

#  t-SNE 
tsne = TSNE(n_components=2, perplexity=5, random_state=0)

#  t-SNE 
X_embedded = tsne.fit_transform(X)

# 
fig, ax = plt.subplots(figsize=(6, 4), dpi=150)
ax.set_facecolor('#F6FAFF')
cmap = cm.RdYlGn_r
norm = Normalize(vmin=0, vmax=600)

# 
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels, cmap=cmap, s=180, edgecolor='white', norm=norm)

# 
cbar = plt.colorbar(scatter)
cbar.set_label('')

# 
plt.title('Fine-tuning Features', fontsize=16)
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)

# 
plt.show()
```
