```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import time
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.feature_selection import VarianceThreshold
from utils.data_processing import read_all_features, read_supplementary_features, feature_concat, data_augmentation, contrastive_learning
from utils.visualization import plot_feature_importance, plot_residuals, plot_prediction_comparison
from utils.model_utils import save_results
from models.regression_model import RegressionModel
from sklearn.metrics import mean_squared_error, r2_score
import joblib
import os

# 忽略警告
import warnings
warnings.filterwarnings('ignore')

# 超参数空间
param_grid = {
    'n_estimators': [80],
    'max_depth': [2],
    'min_samples_split': [7]
}

# 初始化指标列表
all_mse = []
all_rmse = []
all_r2 = []
all_execution_times = []

# 结果汇总字典
seed_results = {}

# 主循环
for seed in tqdm(range(0, 5), desc="Processing seeds"):
    print(f"\n{'='*50}")
    print(f"当前处理SEED{seed}...")
    print(f"{'='*50}")
    start_time = time.time()

    # 文件路径
    main_file = f"fuse_data/imagemol_finturn_Features/SMILES-SEED{seed}-Image_finturn.csv"
    supp_files = [
        #f"fuse_data/lianghua/smiles22_lianghua_seed{seed}.csv",
        f"fuse_data/Morder/smiles22-SEED{seed}_Morder.csv",
        f"fuse_data/CDDD/smiles22_seed{seed}_cddd.csv"
    ]

    # 读取数据
    train_x, test_x, _ = read_all_features([main_file])
    train_supp, test_supp, _ = read_supplementary_features(supp_files)

    # 读取全量数据用于标签处理
    all_data = pd.read_csv(main_file)

    # 初始化模型
    model = RegressionModel(random_state=seed)
    model.fit_label_scaler(all_data)
    model.prepare_labels(all_data)

    # 特征拼接
    train_all, test_all = feature_concat(train_x, train_supp, test_x, test_supp)

    # 特征选择
    var_selector = VarianceThreshold(threshold=0.01)
    train_selected = var_selector.fit_transform(train_all)
    test_selected = var_selector.transform(test_all)

    # 数据增强
    train_aug, y_aug = data_augmentation(train_selected, model.train_y, num_augmented_samples=10)

    # 对比学习
    train_aug, y_aug = contrastive_learning(train_aug, y_aug, num_negative_samples=5)

    # 特征缩放
    train_scaled, test_scaled = model.scale_features(train_aug, test_selected)

    # 超参数调优
    grid_search = model.hyperparameter_tuning(train_scaled, y_aug, param_grid)

    # 预测与评估
    test_pred = model.predict(test_scaled)
    metrics = model.evaluate(test_pred)

    # 记录结果
    all_mse.append(metrics['MSE'])
    all_rmse.append(metrics['RMSE'])
    all_r2.append(metrics['R²'])
    all_execution_times.append(time.time() - start_time)

    # 保存当前seed的结果
    seed_results[seed] = {
        'MSE': metrics['MSE'],
        'RMSE': metrics['RMSE'],
        'R²': metrics['R²'],
        'Execution Time': time.time() - start_time,
        'Best Parameters': grid_search.best_params_ if hasattr(grid_search, 'best_params_') else None
    }

    # 将特征选择器添加到模型对象中
    model.var_selector = var_selector

    # 保存模型（包含特征选择器）
    model_save_dir = "saved_models"
    os.makedirs(model_save_dir, exist_ok=True)
    model_path = f"{model_save_dir}/model_seed{seed}.pkl"
    joblib.dump(model, model_path)
    print(f"✅ 模型已保存至: {model_path}")

    # 单独保存特征选择器（可选，作为备份）
    selector_path = f"{model_save_dir}/var_selector_seed{seed}.pkl"
    joblib.dump(var_selector, selector_path)
    print(f"✅ 特征选择器已保存至: {selector_path}")

    # 可视化与保存
    if seed == 0:
        feature_names = [f"Feature_{i}" for i in range(train_scaled.shape[1])]
        plot_feature_importance(model.best_model, feature_names)
    plot_residuals(model.test_y_original, test_pred)
    save_results(model.best_model, model.test_y_original, test_pred, seed)

    # 输出当前seed的结果
    print(f"\nSEED{seed} 结果:")
    print(f"MSE: {metrics['MSE']:.2f}")
    print(f"RMSE: {metrics['RMSE']:.2f}")
    print(f"R²: {metrics['R²']:.4f}")
    print(f"执行时间: {time.time() - start_time:.2f}秒")
    if hasattr(grid_search, 'best_params_'):
        print(f"最佳参数: {grid_search.best_params_}")

# 结果汇总
mean_mse = np.mean(all_mse)
mean_rmse = np.mean(all_rmse)
mean_r2 = np.mean(all_r2)
mean_time = np.mean(all_execution_times)

print(f"\n{'='*50}")
print(f"五次预测平均值:")
print(f"MSE: {mean_mse:.2f} ± {np.std(all_mse):.2f}")
print(f"RMSE: {mean_rmse:.2f} ± {np.std(all_rmse):.2f}")
print(f"R²: {mean_r2:.4f} ± {np.std(all_r2):.4f}")
print(f"平均执行时间: {mean_time:.2f}秒")
print(f"{'='*50}")

# 打印所有seed的详细结果
print("\n所有SEED的详细结果:")
for seed, result in seed_results.items():
    print(f"\nSEED{seed}:")
    print(f"MSE: {result['MSE']:.2f}")
    print(f"RMSE: {result['RMSE']:.2f}")
    print(f"R²: {result['R²']:.4f}")
    print(f"执行时间: {result['Execution Time']:.2f}秒")
    if result['Best Parameters']:
        print(f"最佳参数: {result['Best Parameters']}")

# 绘制预测对比图（最后一次结果）
plot_prediction_comparison(model.test_y_original, test_pred, mean_mse, mean_rmse, mean_r2)
```
